```python
from difflib import SequenceMatcher
def check_item_macth(string1, string2):
    if len(string1) > len(string2):
        a = string1
        b = string2
    else:
        b = string1
        a = string2
    return a.find(b)

def check_item_macth2(string1, string2):
    if SequenceMatcher(None, string1, string1).ratio() >= 0.8:
        return 0
    else:
        return -1
        

def ismatch(first, second):
    first_tokens = first.split()
    second_tokens = second.split()
    
    if len(first_tokens) < len(second_tokens):
        if len(first_tokens) == 1 and check_item_macth(first_tokens[0], second_tokens[0]) == 0:
            return True
        else:
            return False 
    if len(first_tokens) > len(second_tokens):
        if len(second_tokens) == 1 and check_item_macth(first_tokens[0], second_tokens[0]) == 0:
            return True
        else:
            return False 
    
    for i in range(len(first_tokens)):
        index = check_item_macth(first_tokens[i], second_tokens[i])
        if index != 0:
            return False
    return True
```

```python
matches = []
doubts = []
for i in range(len(lista_produtos)):
    if ismatch(lista_produtos[i], lista_prods_filtrada[i]):
        matches.append((i, lista_produtos[i], lista_prods_filtrada[i]))
    else:
        doubts.append((i, lista_produtos[i], lista_prods_filtrada[i]))
        
with open('matches.csv', 'w') as matchfile:
    for item in matches:
        matchfile.write('{}, {}, {} \n'.format(item[0],item[1], item[2]))
with open('doubts.csv', 'w') as doubtfile:
    for item in doubts:
        doubtfile.write('{}, {}, {} \n'.format(item[0],item[1], item[2]))
        
```
